```python
import serial
import time
from dataclasses import dataclass
from collections import deque
```

```python
SERIAL_PORT = '/dev/ttyACM0'
BAUD_RATE = 115200
ser = serial.Serial(SERIAL_PORT, BAUD_RATE, timeout=0.01)
```

```python
@dataclass
class Counters:
    hit: int
    miss: int
```

```python
# miss statistic: it's a queue of 1 second of data, where each miss is represented as a 1 and each hit as a 0. So the sum of the queue gives the number of misses in the last second, and the length of the queue gives the total number of samples in the last second.
misses = deque(maxlen=100)  # assuming we get 100 samples per second
```

```python
counters = Counters(hit=0, miss=0)

for _ in range(20):
    raw_bytes = ser.readline()
    line = raw_bytes.decode('utf-8', errors='ignore').replace('\x00', '').strip()
    if not line:
        print("Queue is empty")
        misses.append(1)
    else:
        print(line, time.time())
        misses.append(0)
        
print(f"Miss rate: {sum(misses) / len(misses):.2%}")
```
