```python
from scipy.spatial.transform import Rotation as R
import numpy as np
```

```python
roll = 45
pitch = 0
yaw = 0
dx = 0
dy = 0
dz = 0
rotation = R.from_euler('XYZ', [roll, pitch, yaw], degrees=True)
tf = np.eye(4)
tf[:3, :3] = rotation.as_matrix()
tf[:3, 3] = [dx, dy, dz]
print(tf)
```

```python
rotation.apply([1, 1, 1])
```

```python
original_pose = np.eye(4)
original_pose[3, :] = 1
final_pose = tf @ original_pose
print(final_pose)
```

```python
tf @ np.array([1, 1, 1, 1])
```
